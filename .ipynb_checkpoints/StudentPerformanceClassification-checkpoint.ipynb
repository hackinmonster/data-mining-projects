## Introduction

The objective of this investigation is to develop a classification model that can predict what kind of grades a student will receive given information about that student relating to their demographics, personal wellbeing, motivations, and academic profile. A high performing and interpretable model can enable educators and policymakers to better understand how these features impact academic performance, promoting the development of new programs, strategies, and interventions to improve overall academic performance. It can also be used in predicting which students are at risk of failing a given course, empowering educators to take measures in supporting these students. 

The specific aims of this analysis are to:
- Accurately predict students at risk of failing with high recall, minimizing the likelihood of missing any struggling students
- Develop an interpretable model to explain which features from the dataset play the biggest role in explaining grade differences

```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
student_performance = fetch_ucirepo(id=320) 
  
# data (as pandas dataframes) 
X = student_performance.data.features 
y = student_performance.data.targets 
```

```python
X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

```python
X = X.dropna()
```

```python
encoder = LabelEncoder()
X['famsize'] = encoder.fit_transform(X['famsize'])
```

```python
X = pd.get_dummies(X, columns=['Mjob','Fjob','reason','guardian'])
```

```python
X['school'] = X['school'].replace({'GP': 1, 'MS': 0})
X['address'] = X['address'].replace({'U': 1, 'R': 0})
X['sex'] = X['sex'].replace({'F': 1, 'M': 0})
X['famsize'] = X['famsize'].replace({'GT3': 1, 'LE3': 0})
X['Pstatus'] = X['Pstatus'].replace({'A': 1, 'T': 0})
X = X.replace({'yes': 1, 'no': 0})
```

```python
X = X.astype(int)
```

```python
feature_names = X.columns
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
y['pass_fail'] = (y['G3'] >= 10).astype(int)
```

```python
def categorize_grades(grade):
    if grade < 10:
        return 0  # Fail
    elif grade < 14:
        return 1  # Sufficient
    elif grade < 16:
        return 2  # Good
    elif grade < 18:
        return 3  # Very Good
    else:
        return 4  # Excellent

y['grade_category'] = y['G3'].apply(categorize_grades)
```

## Modeling with Logistic Regression

The first model we will look at is logistic regression, because its well suited for binary classification and is highly interpretable. The outputted coefficients are a direct indicator of how predictive a feature is for the target. This will allow us to investigate what student features impact whether they will pass or fail, how to what extent.

```python
X_original = X
y_original = y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size = .25, random_state=42)

class_weight = {0: 20, 1: 1}

logReg = LogisticRegression(class_weight=class_weight)

logReg.fit(X_train, y_train)
y_pred = logReg.predict(X_test)

logReg_score = logReg.score(X_test, y_test)
logReg_coef = logReg.coef_[0]
cm = metrics.confusion_matrix(y_test, y_pred)

print(logReg_score)
print(cm)
print(classification_report(y_test, y_pred))
```

Despite the fact that the dataset is less than 15% fail, 70% of our incorrect classifications were fails. Because the classes are so imbalanced, the model is likely just learning to choose pass because it is usually correct, rather than actually learning the difference between pass and fail. It also means the model is rarely able to correctly predict failing students, which is one of the goals for this classification.

To solve this, we'll rebalance the classes and use Synthetic Minority Oversampling Technique (SMOTE) to create more examples of failing students. To avoid overfitting (since synthetic data isn't new information, just variations of existing data), we'll only apply it to the training set. We'll also pay close attention to precision and recall scores. In our case, recall is the most important because we're trying to predict if a student will fail. Since the dataset is small, SMOTE is likely preferable to downsampling for class rebalancing because we can avoid losing information. 

## Upsampling to Address Class Imbalance

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

X, y = X_original, y_original

X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size=0.3, stratify=y['pass_fail'], random_state=42)

smote = SMOTE(sampling_strategy='minority', random_state=42)

class_weight = {0: 10, 1: 1}

X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

logReg = LogisticRegression(class_weight=class_weight)
logReg.fit(X_train_resampled, y_train_resampled)

y_pred = logReg.predict(X_test)

logReg_score = logReg.score(X_test, y_test)
logReg_coef = logReg.coef_[0]

print(logReg.score(X_test, y_test))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

From using SMOTE, we've increased our recall for course fails (class 0) from .3 to .57 at the expense of our accuracy dropping from .89 to .71.

## Upsampling + Downsampling to Address Class Imbalance

Let's try using a hybrid technique, combining oversampling and undersampling to see if we can improve recall for course fails. 

SMOTEtomek = SMOTE + Tomek

SMOTE is used for oversampling, and works by creating synthetic samples of the minority class by interpolating between existing minority class samples. 

Tomek Links are used for undersampling, and work by removing overlapping majority class samples that are very close to the minority class samples, helping to clean the decision boundary and remove borderline noise.

```python
from imblearn.combine import SMOTETomek

X, y = X_original, y_original

smoteomek = SMOTETomek(sampling_strategy='minority', random_state=42)

class_weight = {0: 10, 1: 1}

X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size=0.3, stratify=y['pass_fail'], random_state=42)

X_resampled, y_resampled = smoteomek.fit_resample(X_train, y_train)

logReg = LogisticRegression(class_weight=class_weight)
logReg.fit(X_resampled, y_resampled)

y_pred = logReg.predict(X_test)
logReg_score = logReg.score(X_test, y_test)
logReg_coef = logReg.coef_[0]


print(logReg_score)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

This didn't seem to change our metrics at all.

```python
for feature, coef in zip(feature_names, logReg_coef):
    print(f"{feature}: {coef}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

feature_importance = pd.DataFrame({"Feature": feature_names, "Coefficient": logReg_coef})
feature_importance["Absolute Coefficient"] = feature_importance["Coefficient"].abs()
feature_importance = feature_importance.sort_values(by="Absolute Coefficient", ascending=False)

plt.figure(figsize=(10, 6))

bars = plt.barh(feature_importance["Feature"], feature_importance["Coefficient"], color="skyblue")

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.axvline(0, color='black', linestyle='dashed', linewidth=1)

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}', ha='left', va='center', fontsize=8)

plt.yticks(fontsize=8)
plt.title("Logistic Regression Coefficients")
plt.xlabel("Coefficient Value")

plt.show()
```

## Modeling wiith K-Nearest Neighbors

The next model we'll use is K-Nearest Neighbors, which has high performance for smaller data sets like ours, and its also possible that predictions can be made with simpler decision boundaries and without the need for complex modeling.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler

X, y = X_original, y_original

# Scale data since we are using KNN and don't want features with higher ranges to have more weight arbitrarily.

scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)

smoteomek = SMOTETomek(sampling_strategy='minority', random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size=0.3, stratify=y['pass_fail'], random_state=42)

X_train, y_train = smoteomek.fit_resample(X_train, y_train)

knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
knn_score = knn.score(X_test, y_test)

print(knn_score)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

From using K-Nearest Neighbors, we were able to achieve a recall of .73 and an accuracy of .57, making it effective but just slightly
worse than our logistic regression model.

## Modeling with Gradient Boosting

The last model we'll try is XGBoost, an ensemble model. Its highly predictive and can capture complex patterns in the data, which may be necessary for our predictions here. We can find the best hyperparameters for maximizing recall for course fails, and then use that model to predict if a student will fail.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, recall_score, make_scorer, f1_score
import xgboost as xgb


X, y = X_original, y_original

X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size=0.3, stratify=y['pass_fail'], random_state=42)

smotetomek = SMOTETomek(sampling_strategy='minority', random_state=42)
X_train, y_train = smotetomek.fit_resample(X_train, y_train)

param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.7, 0.8, 1.0]
}

recall_class_0_scorer = make_scorer(f1_score, pos_label=0)

grid_search = GridSearchCV(xgb.XGBClassifier(random_state=42), param_grid, scoring=recall_class_0_scorer, cv=5)
grid_search.fit(X_train, y_train)

print("Best hyperparameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_

best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)

xgb_score = knn.score(X_test, y_test)


print(xgb_score)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

X, y = X_original, y_original

smote = SMOTE(sampling_strategy='minority', random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y['pass_fail'], test_size=0.3, stratify=y['pass_fail'], random_state=42)

X_train, y_train = smote.fit_resample(X_train, y_train)

rf = RandomForestClassifier(n_estimators=50, random_state=42, class_weight='balanced')

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

rf_score = rf.score(X_test, y_test)
print("Random Forest Model Accuracy:", rf_score)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

## Model Evaluation

With the logistic regression model, we were able to get an accuracy of 58% and recall of 80%

We saw that the highest contributing features to passing a class were:

    -School
    -Interest in higher education
    -Having a father as a guardian
    -Their mother being a teacher
    -Their age
And the highest contributors to failing were:

    -Number of past fails
    -Number of absences
    -Weekend alcohol consumption

Some other interesting insights:

    -More free time was associated with failure
    -The father's job being at home was associated with passing, whereas the mother's job being at home was associated with failing
    -Having extra paid classes for the course subject was associated with failing
    -School choice based on reputation was associated with passing


With the K-Nearest Neighbors model, we got an accuracy of 56% and a recall of 73%.

This makes the KNN just slightly worse overall than our logistic regression model.

Finally, the XGBoost and Random Forest models didn't perform well due to the class imbalance, even when we used hyperparameters most suited to increase recall for class 0 (course fail). The same rebalancing techniques were used for XGBoost as were for KNN and linear regression, and rebalancing further would likely lead to extreme overfitting. Neither were able to achieve strong precision for class 0.
